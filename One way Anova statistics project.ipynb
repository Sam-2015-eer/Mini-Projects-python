# Statistics Project

```python
#importing required libraries from python
import pandas as pd
import matplotlib.pyplot as plt
import pingouin as pg
from scipy import stats
```

```python
#importing read files to read out the data from excel
from pandas import read_excel
df = pd.read_excel('StatProject.xlsx')
df
```

```python
df.drop(df.columns[0], axis=1, inplace=True) #Dropping the Student column from the dataframe
```

```python
df.drop(df.columns[3:26], axis=1, inplace=True)  #Dropping all the nan columns from the dataframe using slicing
df.head()
```

```python
df1=df.rename(columns={'Maths':'M','Physics':'P','Biology':'B'},inplace=False) #Renamed the Column name as M,P,B in a new df1
df1.head()
```

```python
df2=df1.melt(var_name='Subject') #Creating a check point and using the melt() function to transpose/restructure the dataframe.
df2.head()                        #all the subjects are now in a single column with the respective score.
```

```python
df2=df2.rename(columns={'value':'Marks'},inplace=False)  #renaming the value column to Marks in the df2 dataframe
df2.head()
```

## levenes test: To check the homogeneity of variance between the group.

```python
stats.levene(df['Maths'],df['Physics'],df['Biology'])
```

As pvalue > 0.05 we can say that "the groups have equal variance assumed" i.e we failed to reject the Null Hypothesis.

# One-Way Analysis

```python
#One way Anova using pingouin library and more detailed than scipy.
aov=pg.anova(data=df2 , dv='Marks', between='Subject',detailed=True)
aov
```

```python
#One way Anova test using scipy.stats and argument as groups we have in our dataset.
stats.f_oneway(df1['M'],df1['P'],df1['B'])
```

```python
pairwise_test=pg.pairwise_tukey(dv='Marks', between='Subject',data=df2)
pairwise_test
#posthoc analysis(only to be performed if pvalue from anova <0.05 or our alternate hypothesis is true)
```
